````python
ax = sns.barplot(data=df, x='data_ref', y='mau')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_xlabel('Data de Referência')
ax.set_ylabel('Inadimplencia')
ax.set_title('Inadimplencia ao Longo do Tempo')
````

````python
ax = sns.barplot(data=df, x='data_ref', y='posse_de_veiculo')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_xlabel('Data de Referência')
ax.set_ylabel('Posse de Veículo')
ax.set_title('Posse de Veículo ao longo do tempo')
````

````python
# Montando uma tabela da quantidade de filhos médio por data

qtd_filhos_media = df.pivot_table(index='data_ref', values='qtd_filhos', aggfunc='mean')
qtd_filhos_media
````

````python
sns.lineplot(data=qtd_filhos_media, x='data_ref', y='qtd_filhos', marker='D')
plt.title('Média da Quantidade de Filhos ao Longo do Tempo')
plt.xlabel('Data de Referência')
plt.ylabel('Média da Quantidade de Filhos')
plt.xticks(rotation=90)
````

````python
# Montando uma tabela da idade média por data

idade_media = df.pivot_table(index='data_ref', values='idade', aggfunc='mean')
idade_media
````

````python
sns.lineplot(data=idade_media, x='data_ref', y='idade', marker='D')
plt.title('Média da Idade ao Longo do Tempo')
plt.xlabel('Data de Referência')
plt.ylabel('Média da Idade')
plt.xticks(rotation=90)
````

````python
# Montando uma tabela do tempo de emprego médio por data

tempo_emprego_media = df.pivot_table(index='data_ref', values='tempo_emprego', aggfunc='mean')
tempo_emprego_media
````

````python
sns.lineplot(data=tempo_emprego_media, x='data_ref', y='tempo_emprego', marker='D')
plt.title('Média do Tempo de Emprego ao Longo do Tempo')
plt.xlabel('Data de Referência')
plt.ylabel('Média do Tempo de Emprego')
plt.xticks(rotation=90)
````

````python
# Montando uma tabela da quantidade média de pessoas por residência por data

qt_pessoas_residencia_media = df.pivot_table(index='data_ref', values='qt_pessoas_residencia', aggfunc='mean')
qt_pessoas_residencia_media
````

````python
sns.lineplot(data=qt_pessoas_residencia_media, x='data_ref', y='qt_pessoas_residencia', marker='D')
plt.title('Média da Quantidade de Pessoas na Residência ao Longo do Tempo')
plt.xlabel('Data de Referência')
plt.ylabel('Média de Pessoas na Residência')
plt.xticks(rotation=90)
````

````python
# Montando uma tabela da renda média por data

renda_media = df.pivot_table(index='data_ref', values='renda', aggfunc='mean')
renda_media
````

````python
sns.lineplot(data=renda_media, x='data_ref', y='renda', marker='D')
plt.title('Média da Renda ao Longo do Tempo')
plt.xlabel('Data de Referência')
plt.ylabel('Média da Renda')
plt.xticks(rotation=90)
````
