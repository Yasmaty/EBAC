1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
````

````python
df = pd.read_csv('previsao_de_renda.csv')
df.head(3)
````

Na figura abaixo é possível verificar a diferença de renda entre pessoas que possuem imóvel para as que possuem apenas veículo, sendo pintado de laranja as pessoas que possuem imóvel e de azul as que não possuem; E também da para perceber a diferença de quem possuem tanto imóvel quanto veículo, sendo as pessoas que não possuam na esquerda e as que possuem na direita.

````python
ax = sns.pointplot(x='posse_de_veiculo', y='renda', hue = 'posse_de_imovel',
                   data=df, dodge=True, ci = 95)
ax.set_xlabel('Posse de Veículo')
ax.set_ylabel('Renda')
ax.set_title('Relação entre Posse de Veículo, Renda e Posse de Imóvel')
````

Comparando com a inadimplencia, é possível verificar que há um declinio se a pessoa tem veículo e imóvel, se possui ambos o declinio é maior.

````python
ax = sns.pointplot(x='posse_de_veiculo', y='mau', hue = 'posse_de_imovel',
                   data=df, dodge=True, ci = 95)
ax.set_xlabel('Posse de Veículo')
ax.set_ylabel('Inadimplencia')
ax.set_title('Relação entre Posse de Veículo, Renda e Posse de Imóvel')
````

Em ambos os exemplos também foi possível verificar que os clientes que possuem apenas veículo tem uma variância maior, o que os tornam mais volateis.
