1. Uma mquina embala rao para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatrio, com distribuio Normal, com mdia 15kg, mas com desvio padro 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg.

> Para resolver esta primeira questo,  necessrio usar a Funo de Distribuio Acumulativa (**F(X)**), responsvel por medir a probabilidade de uma varivel aleatria ser menor do que um determinado valor (**P(X < x)**):
>
> $$ F(X < x) = P(X < x)  $$
>
> Considerando que queremos saber a probabilidade do pacote selecionado ter mais do que 15,2 kg e que, em uma distribuio Normal, o intervalo de valores  infinito, precisamos fazer o clculo:
>
> $$ P(X > 15,2) = F(\infty) - F(15,2)$$ 
>
> O clculo desta probabilidade encontra-se abaixo.

```python
# Mdia "mu" e Desvio Padro "sigma"
mu = 15
sigma = 0.1

# Distribuio Normal "N" com a Mdia de Desvio Padro informados
N = stats.norm(loc=mu, scale=sigma)

# Valores mnimo "low" e mximo "high"
low = 15.2 
high = np.inf

# Probabilidade do peso maior que 15.2 kg
prob = N.cdf(high) - N.cdf(low)

print(f'A probabilidade para valores maiores que {low:.1f} kg  de {prob:.2%}'.replace('.',',')+'.')
```

2. Da mquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

> Para calcular o percentil de ordem 95%, precisamos usar a Funo Ponto Percentil, sendo esta a inversa da CDF. Ou seja, ela devolve o valor mximo da varivel aleatria, para que uma seleo de qualquer valor menor do que o mximo tenha probabilidade na ordem da informada. 
> Na clula abaixo encontra-se o clculo do peso dos pacotes.

```python
# Calcula a Funo Ponto Percentil para 95% para o peso dos pacotes.
ppf = N.ppf(.95)

# Imprime o peso dos pacotes que obedece  PPF(95%)
print(f'H 95% de chances de se obter pacotes com peso inferior a {ppf:.2f} kg'.replace('.',',') + '.')
```

3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

> Para que sejam calculadas as variveis aleatrias para cada valor do objeto **Y**, vamos criar dois vetores: **Y1** e **Y2**, sendo que o primeiro tem distribuio Normal padro, enquanto que o segundo, distribuio t-student. 

```python
# Vetor com os valores do objeto Y (valores entre -6 e 6)
Y=np.linspace(-6, 6, 200)

# Parmetros da distribuio Normal do vetor Y1 (mdia, varincia e Desvio Padro)
mu_Y1 = 0
var_Y1 = 1
sigma_Y1 = math.sqrt(var_Y1)

# Vetor Y1 com os valores da Funo de Distribuio Acumulativa para todo o Y
Y1 = stats.norm.cdf(Y, mu_Y1, sigma_Y1) 

# Imprime o vetor Y1
Y1
```

```python
# Cria o vetor Y2 com 5 graus de liberdade, calculando a CDF para cada valor de Y  
Y2 = stats.t.cdf(Y, 5) 

# Imprime o vetor Y2
Y2
```

4. Faa uma visualizao grfica dos dados do item anterior.

> Agora que os vetores **Y1** e **Y2** esto prontos, vamos fazer o grfico de seus valores. Para compar-los, vamos imprimir ambos no mesmo grfico.

```python
# Cria uma figura com largura = 10 e altura = 4 para o grfico
fig, ax = plt.subplots(figsize=(10,4))

# Plota o vetor Y1 em funo de Y, rotulado por "Y1 - Normal"
ax.plot(Y, Y1, label='Y1 - Normal')

# Plota o vetor Y2 em funo de Y, rotulado por "Y2 - t(5)"
ax.plot(Y, Y2, label='Y2 - t(5)')

# Insere a legenda (rtulo) no grfico
ax.legend()
```

> O grfico mostra que valores da CDF em ambas as distribuies so muito prximos um do outro, sendo que a t-student apresenta valores superiores  distribuio Normal para Y < 0, e o inverso para Y > 0.
