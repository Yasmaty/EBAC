### <font color="blue"> Frmula de Bhaskara </font> 
A soluo para uma funo de segundo grau depende das suas razes (valores de x). Como j vimos, os coeficientes precisam ser nmeros reais e o angular diferente de zero. Sendo assim, temos a seguinte frmula de Bhaskara:



# $$ x=\frac{-b \pm \sqrt{b^2 - 4 \cdot a \cdot c}}{2 \cdot a}$$
<div style='text-align: center;'>
    <strong><em>Frmula de Bhaskara.</em></strong>
</div>



Para melhor compreenso o seu clculo  dividido em duas partes: discriminante da equao e operaes para determinar as razes.


<strong> Discriminante </strong>

A figura dentro da raiz na frmula de Bhaskara  nomeada de discriminante. Seu smbolo  a letra grega delta e apresenta a determinada frmula:

# $$ \Delta = b^2 - 4 \cdot a \cdot c $$
<div style='text-align: center;'>
   <strong><em>Frmula da discriminante.</em></strong>
</div>


- Se o delta for maior que zero, a equao ter dois valores reais e distintos.  
- Se o delta for igual a zero, a equao ter somente um valor real ou dois resultados iguais. 
- Se o delta for menor que zero, a equao no possuir valores reais.
