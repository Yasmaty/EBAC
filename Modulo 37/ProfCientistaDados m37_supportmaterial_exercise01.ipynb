1. Qual a principal diferença entre um *behaviour score* e um *credit score*?

> No _credit score_ é avaliado o risco de crédito que um cliente pode apresentar por meio da utilização de métodos estatísticos e probabilísticos que resultam em uma determinada pontuação (_score_). A partir desse _score_, a organização em questão (bancos por exemplo) decidem se vão aceitar o indivíduo como cliente ou não.
> Por outro lado, no _behaviour scoring_ é avaliado o perfil comportamental que um cliente tem com a organização, ou seja, se ele apresenta ou não atrasos no pagamento, se vale a pena conceder maior limite em operações de crédito ou alterar juros. 

2. Por que é tão comum se estabelecer janelas de desempenho de 12 meses, e 12 safras para o desenvolvimento de modelos?

> O uso de janelas de desempenho de 12 meses e 12 safras para o desenvolvimento de modelos tem por objetivo eliminar efeitos de sazonalidade, ou seja eventos/comportamentos que ocorreram por um breve período de tempo capazes de alterar errôneamente o resultado do modelo (como classificar um "bom" pagador como "mau" devido a um curto atraso no pagamento).

3. Qual a diferença entre amostra *out of time* e amostra *out of sample*?

> Amostras _out of time_ são safras cujo período é posterior ao usado para o desenvolvimento e validação de um modelo (por exemplo, dados do ano de 2022 utilizados em um modelo validado com dados do ano de 2021).
> 
> Amostra _out of sample_ são as safras utilizadas para validação do modelo.

4. Se os dados da variavel resposta estão sistematicamente corrompidos para duas das safras de desenvolvimento, o que você faria?

> Se for possível, eliminaria as duas safras em questão, caso contrário procuraria por técnicas capazes de corrigirem os erros presentes nestas safras.
